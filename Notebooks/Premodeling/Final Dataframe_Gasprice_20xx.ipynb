#### The main goal of our work is to use data of gas consumption and gas prices from the past 4 years (2018 - 2021) to predict consumption and prices for 2022, as if the Ukraine war would have not happened.
#### We will compare this predicted data to the actual data (01 - 07/2022) to show the effect of the war on the consumption and the prices.
#### As the last step, we will predict gas consumption and prices based on the current data (01 - 07/2022) with diffent scenarios of the gas supply by Russia.

### **This notebook will concatenate all data frames into one to do the modeling on that.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_price_20xx = pd.read_csv(r"/Volumes/Extreme Pro/WBS Bootcamp Data Science/Final Project/Data/Cleaned/Gas prices/Final Data_Gas price EU_2018-2021_daily.csv")
data_supply_20xx = pd.read_csv(r"/Volumes/Extreme Pro/WBS Bootcamp Data Science/Final Project/Data/Cleaned/Gas supply/Final Data_Gas supply_2018-2021_daily.csv")
data_infl_20xx = pd.read_csv(r"/Volumes/Extreme Pro/WBS Bootcamp Data Science/Final Project/Data/Cleaned/Inflation/Final Data_Inflation EU_2018-2021_monthly.csv")
data_cons_20xx = pd.read_csv(r"/Volumes/Extreme Pro/WBS Bootcamp Data Science/Final Project/Data/Cleaned/Gas consumption/Final Data_Consumption_2018-2021_daily.csv")
```

```python
data_price_20xx = data_price_20xx.drop(data_price_20xx.columns[0], axis=1)
data_price_20xx
```

```python
data_supply_20xx = data_supply_20xx.drop(data_supply_20xx.columns[0], axis=1)
data_supply_20xx
```

```python
data_infl_20xx = data_infl_20xx.drop(data_infl_20xx.columns[0], axis=1)
data_infl_20xx
```

```python
data_cons_20xx
```

```python
data_cons_20xx = data_cons_20xx.drop(data_cons_20xx.columns[[3, 4, 5, 6]], axis = 1)
data_cons_20xx
```

```python
# merge supply with consumption
final_df_gasprice_20xx = data_supply_20xx.merge(data_cons_20xx, how="left", on = ["day", "month", "year"])
final_df_gasprice_20xx
```

```python
# merge with price
final_df_gasprice_20xx = final_df_gasprice_20xx.merge(data_price_20xx, how="left", on = ["day", "month", "year"])
final_df_gasprice_20xx
```

```python
final_df_gasprice_20xx = final_df_gasprice_20xx.drop(final_df_gasprice_20xx.columns[6], axis=True)
final_df_gasprice_20xx.rename(columns={final_df_gasprice_20xx.columns[2]: "month_no"}, inplace=True)
final_df_gasprice_20xx
```

```python
# merge with inflation
final_df_gasprice_20xx = final_df_gasprice_20xx.merge(data_infl_20xx, how="left", on = ["month", "year"])
final_df_gasprice_20xx
```

```python
final_df_gasprice_20xx = final_df_gasprice_20xx.drop(final_df_gasprice_20xx.columns[7], axis=True)
final_df_gasprice_20xx.rename(columns={final_df_gasprice_20xx.columns[2]: "month_no"}, inplace=True)
final_df_gasprice_20xx
```

```python
final_df_gasprice_20xx.loc[len(final_df_gasprice_20xx.index)] = [31, "December", 12, 2017, 150, 4000, 1, 19.62]
final_df_gasprice_20xx
```

```python
final_df_gasprice_20xx = final_df_gasprice_20xx.sort_values(["year", "month_no", "day"]).reset_index(drop=True)
final_df_gasprice_20xx
```

```python
final_df_gasprice_20xx = final_df_gasprice_20xx.interpolate()
final_df_gasprice_20xx
```

```python
final_df_gasprice_20xx.isna().sum()
```

```python
final_df_gasprice_20xx.info()
```

```python
final_df_gasprice_20xx = final_df_gasprice_20xx.drop(final_df_gasprice_20xx.index[0], axis=0)
final_df_gasprice_20xx
```

```python
final_df_gasprice_20xx.to_csv("Final DF_gasprice_2018-2021.csv")
```
