#### The main goal of our work is to use data of gas consumption and gas prices from the past 4 years (2018 - 2021) to predict consumption and prices for 2022, as if the Ukraine war would have not happened.
#### We will compare this predicted data to the actual data (01 - 07/2022) to show the effect of the war on the consumption and the prices.
#### As the last step, we will predict gas consumption and prices based on the current data (01 - 07/2022) with diffent scenarios of the gas supply by Russia.

### **This notebook will apply the different scenario of Russian gas supply onto the gas supply dataframe.**

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
data_supply_2021 = pd.read_csv(r"/Volumes/Extreme Pro/WBS Bootcamp Data Science/Final Project/Data/Cleaned/Gas supply/Final Data_Gas supply_2018-2021_daily.csv")
data_supply_2021
```

```python
data_supply_2021 = data_supply_2021.drop(data_supply_2021.columns[0], axis=1)
data_supply_2021
```

```python
data_supply_2022 = pd.read_csv(r"/Volumes/Extreme Pro/WBS Bootcamp Data Science/Final Project/Data/Cleaned/Gas supply/Final Data_Gas supply_2022_daily.csv")
data_supply_2022
```

```python
data_supply_2022 = data_supply_2022.drop(data_supply_2022.columns[0], axis=1)
data_supply_2022
```

```python
# selecting rows from May 2021 to December 2021
data_supply_2021 = data_supply_2021.query("year == 2021")
data_supply_2021
```

```python
data_supply_2021 = data_supply_2021.loc[data_supply_2021["month"].isin(["May", "June", "July", "August", "September", "October", "November", "December"])]
data_supply_2021
```

```python
data_supply_12mon = pd.concat([data_supply_2021, data_supply_2022])
data_supply_12mon = data_supply_12mon.reset_index(drop=True)
data_supply_12mon
```

```python
sns.set(rc={'figure.figsize':(20,10)})
sns.lineplot(data=data_supply_12mon, x="month", y="Gas supply by Russia");
```

## Scenario 1 = Russia supplies same amount of gas as in last 12 months (May 2021 - April 20022)

```python
data_supply_scen1 = pd.DataFrame(data_supply_12mon["year"]+1)
data_supply_scen1.rename(columns={data_supply_scen1.columns[0]: "new_year"}, inplace=True)
data_supply_scen1
```

```python
data_supply_scen1 = data_supply_scen1.join(data_supply_12mon)
data_supply_scen1
```

```python
data_supply_scen1 = data_supply_scen1.drop(data_supply_scen1.columns[4], axis = 1)
data_supply_scen1.rename(columns={data_supply_scen1.columns[0]: "year"}, inplace=True)
data_supply_scen1 = data_supply_scen1[["day", "month", "month_no", "year", "Gas supply by Russia"]]
data_supply_scen1
```

```python
data_supply_scen1.to_csv("Final DF_Supply_Scen1_daily.csv")
```

## Scenario 2 = Russia supplies 50% of amount of gas as in last 12 months (May 2021 - April 20022)

```python
data_supply_scen2 = pd.DataFrame(data_supply_scen1["Gas supply by Russia"]*0.5)
data_supply_scen2
```

```python
data_supply_scen2 = data_supply_scen2.join(data_supply_scen1.drop(data_supply_scen1.columns[4], axis=1))
data_supply_scen2 = data_supply_scen2[["day", "month", "month_no", "year", "Gas supply by Russia"]]
data_supply_scen2
```

```python
data_supply_scen2.to_csv("Final DF_Supply_Scen2_daily.csv")
```
